```python
import numpy as np
import matplotlib.pyplot as plt

# Функція, що моделює сигнал
def generate_signal(timesteps, frequency, amplitude):
    return amplitude * np.sin(2 * np.pi * frequency * timesteps)

# Функція, що додає шум до сигналу
def add_noise(signal, noise_level):
    noise = noise_level * np.random.randn(len(signal))
    return signal + noise

# Функція для обчислення вихідного сигналу з використанням стохастичного резонансу
def stochastic_resonance(input_signal, noise_level, threshold):
    return (input_signal + noise_level * np.random.randn(len(input_signal))) > threshold

# Функція для обчислення SNR
def calculate_snr(signal, noise):
    signal_power = np.mean(signal ** 2)
    noise_power = np.mean(noise ** 2)
    return 10 * np.log10(signal_power / noise_power)

# Функція для обчислення MSE
def calculate_mse(signal, filtered_signal):
    return np.mean((signal - filtered_signal) ** 2)

# Функція для візуалізації сигналу
def plot_signal(signal, title):
    plt.plot(signal)
    plt.title(title)
    plt.xlabel('Час')
    plt.ylabel('Амплітуда')
    plt.show()

# Параметри сигналу і обробки
timesteps = np.linspace(0, 1, 1000)  # Часові відліки
frequency = 10  # Частота сигналу
amplitude = 1  # Амплітуда сигналу
noise_level = 0.5  # Рівень шуму
threshold = 0  # Поріг для стохастичного резонансу

# Генерація сигналу
signal = generate_signal(timesteps, frequency, amplitude)

# Додавання шуму до сигналу
noisy_signal = add_noise(signal, noise_level)

# Візуалізація сигналу перед обробкою
plot_signal(noisy_signal, 'Сигнал перед обробкою')

# Обчислення SNR до обробки
snr_before = calculate_snr(signal, noisy_signal)
print("Сигнал-шум співвідношення (SNR) до обробки:", snr_before)

# Використання стохастичного резонансу
output_signal = stochastic_resonance(noisy_signal, noise_level, threshold)

# Візуалізація сигналу після обробки
plot_signal(output_signal, 'Сигнал після стохастичного резонансу')

# Обчислення SNR після обробки
snr_after = calculate_snr(signal, output_signal)
print("Сигнал-шум співвідношення (SNR) після обробки:", snr_after)

# Обчислення MSE
mse = calculate_mse(signal, output_signal)
print("Середньоквадратична помилка (MSE):", mse)
```
