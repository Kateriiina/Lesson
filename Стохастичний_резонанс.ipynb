```python
import numpy as np
import matplotlib.pyplot as plt

# ,   
def generate_signal(timesteps, frequency, amplitude):
    return amplitude * np.sin(2 * np.pi * frequency * timesteps)

# ,     
def add_noise(signal, noise_level):
    noise = noise_level * np.random.randn(len(signal))
    return signal + noise

#         
def stochastic_resonance(input_signal, noise_level, threshold):
    return (input_signal + noise_level * np.random.randn(len(input_signal))) > threshold

#    SNR
def calculate_snr(signal, noise):
    signal_power = np.mean(signal ** 2)
    noise_power = np.mean(noise ** 2)
    return 10 * np.log10(signal_power / noise_power)

#    MSE
def calculate_mse(signal, filtered_signal):
    return np.mean((signal - filtered_signal) ** 2)

#    
def plot_signal(signal, title):
    plt.plot(signal)
    plt.title(title)
    plt.xlabel('')
    plt.ylabel('')
    plt.show()

#    
timesteps = np.linspace(0, 1, 1000)  #  
frequency = 10  #  
amplitude = 1  #  
noise_level = 0.5  #  
threshold = 0  #    

#  
signal = generate_signal(timesteps, frequency, amplitude)

#    
noisy_signal = add_noise(signal, noise_level)

#    
plot_signal(noisy_signal, '  ')

#  SNR  
snr_before = calculate_snr(signal, noisy_signal)
print("-  (SNR)  :", snr_before)

#   
output_signal = stochastic_resonance(noisy_signal, noise_level, threshold)

#    
plot_signal(output_signal, '   ')

#  SNR  
snr_after = calculate_snr(signal, output_signal)
print("-  (SNR)  :", snr_after)

#  MSE
mse = calculate_mse(signal, output_signal)
print("  (MSE):", mse)
```
