```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
planets = sns.load_dataset("planets")
```

```python
print("Загальна інформація про набір даних:")
pl_data = pd.DataFrame(planets)
pl_data.info()
```

```python
print("\nРозмір набору даних:")
pl_data.shape
```

```python
print("\nОсновні статистики за набором даних:")
pl_data.describe
```

```python
print("\nПерші сім рядків:")
pl_data.head(7)
```

```python
print("\nОстанні сім рядків:")
pl_data.tail(7)
```

```python
print("\nЗначення ознак method і year:")
print("Method:")
pl_data['method'].unique()
pl_data['method'].value_counts()
```

```python
print("\nYear:")
pl_data['year'].unique()
pl_data['method'].value_counts()
```

```python
print("\nЗаписи, які відповідають 2003 року:")
pl_data[planets['year'] == 2003]
```

```python
max_mass_object = pl_data[pl_data['mass'] == pl_data['mass'].max()]
print("\nОб'єкт з максимальною масою:")
print(max_mass_object)
```

```python
min_orbital_period = pl_data['orbital_period'].min()
print("\nМінімальний орбітальний період:", min_orbital_period)
```

```python
grouped_data = pl_data.groupby('method').agg({'orbital_period': ['min', 'max'], 'mass': ['min', 'max'], 'distance': ['min', 'max']})
print("\nМінімальне і максимальне значення ознак за кожним method:")
print(grouped_data)
```
